```python
import torch.nn.functional as F

def generate(n_words=20):
    context = [0] * words_context  # start with all zeros (same as training)
    result = []
    for _ in range(n_words):
        emb = C[torch.tensor([context])]
        h = torch.tanh(emb.view(1, -1) @ w1 + b1)
        logits = h @ w2 + b2
        probs = F.softmax(logits, dim=1)
        ix = torch.multinomial(probs, num_samples=1).item()
        result.append(itow[ix])
        context = context[1:] + [ix]
    return ' '.join(result)

# Example usage
print(generate(100))
```

```python

import torch.nn.functional as F

# sample sentences from the trained word-level model
g = torch.Generator().manual_seed(99 + 1)  # reproducibility

for _ in range(20):  # generate 20 sentences
    out = []
    context = [0] * words_context  # start with special "start" context
    while True:
        emb = C[torch.tensor([context])]               # (1, context, embedding_size)
        h = torch.tanh(emb.view(1, -1) @ w1 + b1)      # hidden layer
        logits = h @ w2 + b2                           # output layer
        probs = F.softmax(logits, dim=1)               # convert to probabilities
        ix = torch.multinomial(probs, num_samples=1, generator=g).item()  # sample next word

        context = context[1:] + [ix]                   # slide window
        out.append(ix)

        if itow[ix] == '.':                            # stop at end-of-sentence token
            break

    print(' '.join(itow[i] for i in out))
```

```python
import torch
import torch.nn.functional as F

def most_similar(word, top_k=5):
    if word not in wtoi:
        print(f"'{word}' not found in vocabulary.")
        return []

    # get the embedding of the given word
    word_emb = C[wtoi[word]]  # shape: (embedding_size,)

    # compute cosine similarity with all embeddings
    similarities = F.cosine_similarity(word_emb.unsqueeze(0), C, dim=1)

    # get top_k most similar indices (excluding itself)
    topk = torch.topk(similarities, top_k + 1)  # +1 to skip the word itself
    similar_indices = [i.item() for i in topk.indices if i.item() != wtoi[word]][:top_k]

    # return the most similar words and their similarity scores
    results = [(itow[i], similarities[i].item()) for i in similar_indices]

    return results

# Example usage:
word = "Garba"   # replace with any word from your vocab
similar = most_similar(word, top_k=5)
for w, score in similar:
    print(f"{w:15s}  similarity={score:.3f}")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def fun(x):
  return (x-4) *(x-5) +9
```

```python
def derivative(x):
  return 2*x-9
```

```python
def gradient_descent(iterations,learning_rate,x):
  x_values = []
  func_values = []
  for i in range(iterations):
    x_values.append(x)
    func_values.append(fun(x))
    x = x-learning_rate*(derivative(x))
  plt.plot(x_values,func_values)

```

```python
points = []
values = []
for i in np.arange(-2.5,10,0.1):
  points.append(i)
  values.append(fun(i))
plt.plot(points,values)
gradient_descent(100,0.03,-2)
```
